```python
import easyocr
import gradio as gr

print(easyocr.__version__)
print(gr.__version__)
```

```python
def sample_ocr(image):
    reader = easyocr.Reader(['ko','en'])
    results = reader.readtext(image)
    return results
```

```python
demo = gr.Interface(
    sample_ocr,
    gr.Image(type="filepath"),
    outputs="text"
)
demo.launch()
```
